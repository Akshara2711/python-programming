```python

x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(2, 3, figsize=(12, 6))
for i in range(2):
    for j in range(3):
        y = np.sin(x + (i * 3 + j))  
        ax = axes[i, j]
        ax.plot(x, y)
        ax.set_title(f'plot{i * 3 + j + 1}')
        ax.grid(True)
fig.suptitle('graph', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

```python


import numpy as np
import matplotlib.pyplot as plt
groups = ['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5']
men_scores = np.array([22, 30, 35, 35, 26])
women_scores = np.array([25, 32, 30, 35, 29])
x = np.arange(len(groups)) 
width = 0.35  


a,ax = plt.subplots()
ax.bar(x - width/2, men_scores, width, label='Men', color='blue')
ax.bar(x + width/2, women_scores, width, label='Women', color='pink')

ax.set_xlabel('Groups')
ax.set_ylabel('Scores')
ax.set_title('Scores of Men and Women Across Five Groups')
ax.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_csv('company_sales_data.csv')
months = data['month_number']
face_wash_sales = data['facewash']
face_cream_sales = data['facecream']
bathing_soap_sales = data['bathingsoap']
bar_width = 0.4
x_indexes = np.arange(len(months))

plt.figure(figsize=(10,5))
plt.bar(x_indexes - bar_width/2, face_wash_sales, width=bar_width, label='Face Wash', color='blue')
plt.bar(x_indexes + bar_width/2, face_cream_sales, width=bar_width, label='Face Cream', color='green')

plt.xlabel('Month')
plt.ylabel('Sales Data')
plt.title('Face Wash vs Face Cream Sales Per Month')
plt.xticks(ticks=x_indexes, labels=months)
plt.legend()
plt.show()
plt.figure(figsize=(10,5))
plt.bar(months, bathing_soap_sales, color='cyan')

plt.xlabel('Month')
plt.ylabel('Sales Data')
plt.title('Bathing Soap Sales Per Month')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('company_sales_data.csv')
products = ['facewash', 'facecream', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total_sales = [data[product].sum() for product in products]
plt.figure(figsize=(8,8))
plt.pie(total_sales, labels=products, autopct='%1.1f%%', colors=['blue', 'green', 'red', 'cyan', 'purple', 'orange'])

plt.title('Total Sales Distribution of Products')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('company_sales_data.csv')
months = data['month_number']
bathing_soap_sales = data['bathingsoap']
face_wash_sales = data['facewash']
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(months, bathing_soap_sales, marker='o', linestyle='-', color='c')
plt.xlabel('Month')
plt.ylabel('Sales Data')
plt.title('Bathing Soap Sales Per Month')
plt.subplot(1, 2, 2)
plt.plot(months, face_wash_sales, marker='s', linestyle='-', color='b')
plt.xlabel('Month')
plt.ylabel('Sales Data')
plt.title('Face Wash Sales Per Month')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('company_sales_data.csv')

months = data['month_number']
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
sales_data = [data[product] for product in products]

plt.figure(figsize=(10,6))
plt.stackplot(months, sales_data, labels=products, colors=['blue', 'green', 'red', 'cyan', 'purple', 'orange'])

plt.xlabel('Month')
plt.ylabel('Sales Data')
plt.title('Stacked Area Plot of Product Sales')
plt.legend(loc='upper left')
plt.show()
```
